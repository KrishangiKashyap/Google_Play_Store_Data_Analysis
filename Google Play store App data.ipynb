### Converting Object types to numeric values

```python
df['Size'] = df['Size'].apply((lambda x: str(x).replace('M', '') if 'M' in str(x) else str(x)))
df['Size'] = df['Size'].apply((lambda x: str(x).replace('k', '') if 'k' in str(x) else str(x)))
df['Size'] = df['Size'].apply((lambda x: str(x).replace(',', '') if ',' in str(x) else str(x)))
df['Size'] = df['Size'].apply((lambda x: str(x).replace('G', '') if 'G' in str(x) else str(x)))

substring = 'Varies with device'
filter = df['Size'].str.contains(substring)
filtered_df = df[~filter]

filtered_df['Size'] = filtered_df['Size'].apply((lambda x: float(x)))
```

```python
filtered_df['Size']
```

```python
grp1 = filtered_df.groupby('Category')
D = grp1['Size'].agg(np.mean)

plt.figure(figsize=(16,5))
plt.plot(D, 'r--', color='blue')
plt.xticks(rotation=90)
plt.title('Category vs Size of Apps')
plt.xlabel('Categories')
plt.ylabel('Size of Apps')
plt.show()
```
